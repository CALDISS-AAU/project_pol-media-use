```python
import pandas as pd
import json
import os
from os import listdir
from os.path import isfile, join
import re
```

```python
datapath = os.path.join('D:/', 'data',  'dk_news', 'articles_20210421_with-text')

datafiles = [join(datapath,f) for f in listdir(datapath) if isfile(join(datapath, f))]

datafiles = [datafile for datafile in datafiles if re.match(r'.*\.json', datafile)]
```

```python
articles_df = pd.DataFrame()

for c, datafile in enumerate(datafiles, start = 1):
    articles_df = articles_df.append(pd.read_json(datafile), ignore_index = True)
    
    print("{:.2f}%".format(100.0*c/len(datafiles)), end = "\r")
    
articles_df.head()
```

```python
articles_df.columns
```

```python
articles_df['article_datetime'] = pd.to_datetime(articles_df['article_datetime'], utc = True)
```

```python
filename = "articles_combined_20210305.zip"
fileout = os.path.join(datapath, filename)
articles_df.to_csv(fileout, index = False, compression = "zip")
```

```python
articles_sample = articles_df.groupby('newspaper_name').sample(100)
```

```python
filename = "articles_combined_sample_20210305.zip"
fileout = os.path.join(datapath, filename)
articles_sample.to_csv(fileout, index = False, compression = "zip")
```
